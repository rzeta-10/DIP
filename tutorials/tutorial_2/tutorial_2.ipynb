```python
img = cv2.imread("scenary.jpeg", cv2.IMREAD_COLOR)

img_rgb = img[:, :, ::-1]
plt.imshow(img_rgb)
```

```python
crop_region = img[0:50,100:150]
plt.imshow(crop_region)
```

```python
resize_crop_img = cv2.resize(crop_region,None,fx=2,fy=2) #fx , fy are scaling factors
plt.imshow(resize_crop_img)
```

```python
#for specific width and height
desired_h = 100
desired_w = 80

dim = (desired_w, desired_h)

resize_crop_img = cv2.resize(crop_region, dsize = dim, interpolation = cv2.INTER_AREA)
plt.imshow(resize_crop_img)
```

```python
# resizing and maintaining aspect ratio
desired_w = 100
aspect_ratio = desired_w / crop_region.shape[1]
desired_h = int(crop_region.shape[0] * aspect_ratio)

dim = (desired_w, desired_h) # y, x

resize_crop_img = cv2.resize(crop_region, dsize = dim, interpolation = cv2.INTER_AREA)
plt.imshow(resize_crop_img)
```

```python
resize_crop_img = resize_crop_img[:, : , ::-1]

cv2.imwrite("resize_crop_img.png",resize_crop_img)
plt.imshow(resize_crop_img)
```

```python
# flip -> 0 == around x-axis , +ve == around y-axis , -ve == both

pic = cv2.imread("naruto_pixel.png")
pic = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)
pic_horz = cv2.flip(pic,1)
pic_vert = cv2.flip(pic,0)
pic_both = cv2.flip(pic,-1)

plt.figure(figsize=[20,10])
plt.subplot(141); plt.imshow(pic_horz); plt.title("Horizontal Flip")
plt.subplot(142); plt.imshow(pic_vert); plt.title("Vertical Flip")
plt.subplot(143); plt.imshow(pic_both); plt.title("Both Flip")
plt.subplot(144); plt.imshow(pic); plt.title("Original")
```
