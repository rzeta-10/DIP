```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread("gojo_pixel.jpeg")

plt.imshow(img[:,:,::-1])
```

```python
img_cpy = img.copy()

cv2.line(img_cpy, (20, 10), (50,10), (255,0,0), thickness=3, lineType=cv2.LINE_AA)

plt.imshow(img_cpy)
```

```python
img_cpy = img.copy()

cv2.circle(img_cpy, (75,160), 25, (0,0,255), thickness=5, lineType=cv2.LINE_AA) # if thickness is negative , then the circle is filled inside
plt.imshow(img_cpy)
```

```python
img_cpy = img.copy()

cv2.rectangle(img_cpy, (60,175), (100,150), (0,255,0), thickness = 1, lineType=cv2.LINE_AA)
plt.imshow(img_cpy)
```

```python
img_text = img.copy()
text = "Chibi Gojo in Pixel Art"
fontScale = 1 # scale -ve => upside down text
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (0,0,255)
fontThickness = 2

cv2.putText(img_text, text, (20,230), fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA) # LINE_4, LINE_8
plt.imshow(img_text[:,:,::-1])
```
